## Exploratory Data Analysis of Electric_Vehicle_Population




### 1. **Dataset Introduction:**

This dataset contains information related to electric vehicles (EVs) and their characteristics. It is a comprehensive collection of data that we will be leveraging for our data science project.

**Key Information:**

- **Size:** The dataset comprises 117,696 records and 17 columns, making it a substantial resource for our analysis.

- **Columns:** The dataset includes various columns providing details about each electric vehicle, such as its make, model, and electric range.

- **Missing Values:** Notably, there are missing values in the "Clean Alternative Fuel Vehicle (CAFV) Eligibility" column, which we will address during our data analysis.

- **Data Types:** We have a mix of data types, with most columns being categorical (e.g., make and model) and others containing numerical data (e.g., electric range).

**Objective:**

Our primary goal with this dataset is to predict the electric range of these vehicles. Electric range is a crucial factor for EV enthusiasts and potential buyers. By understanding the factors that influence electric range, we can provide valuable insights for both the EV industry and consumers.

### Analyzing Variables

```python
import matplotlib.pyplot as plt

plt.hist(df['Electric Range'], bins = 20, color = 'blue', edgecolor = 'black')
plt.xlabel('Electric Range')
plt.ylabel('Frequency')
plt.title('Histogram of Electric Range')
plt.show()
```


**Interpretation:**

The histogram depicts the distribution of electric range values. The x-axis represents the electric range, and the y-axis represents the frequency of occurrences for each range. The histogram exhibits a right-skewed distribution, indicating that a higher concentration of values falls within the lower electric range. The peak of the histogram lies around 0.2, suggesting that the most frequent electric range value is around 0.2.

**Observations:**

* The electric range values are predominantly concentrated between 0.0 and 0.4.
* The frequency of occurrences gradually decreases as the electric range increases.
* There are relatively few occurrences for electric range values exceeding 0.8.

### 6. Multivariate Analysis:


```python
# Select only numerical columns for correlation matrix
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Pairwise correlation matrix
correlation_matrix = df[numerical_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Pairwise Correlation Heatmap')
plt.show()
```
